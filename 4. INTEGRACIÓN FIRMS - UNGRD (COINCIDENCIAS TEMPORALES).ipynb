## 4.1 Preprocesar y Filtrar Base de Datos de la Unidad Nacinal de Gestión del Riesgo de Desastre por Fecha, Evento y Departamento ##

Preprocesamiento, diagnóstico y filtrado de una base de datos de eventos de la UNGRD (Unidad Nacional para la Gestión del Riesgo de Desastres) almacenada en formato Excel. Se identifican y extraen registros de incendios ocurridos en departamentos de interés (Boyacá, Santander y Cundinamarca), incorporando un control temporal detallado (año, mes y día) y generando resúmenes diagnósticos por año.
