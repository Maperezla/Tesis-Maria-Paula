# 5. GENERACI√ìN DE PUNTOS DE AUSENCIA DE INCENDIOS #

Se genera un conjunto de puntos aleatorios ‚Äúno calientes‚Äù (ausencias) dentro del √°rea de inter√©s, asegurando que dichos puntos no est√©n cerca de los puntos FIRMS (presencias). ## 5.1  Generaci√≥n de Puntos No Calientes Aleatorios Fuera del Buffer de 1 km de Puntos Calientes ##

Carga de librerias, lectura del √°rea de interes y puntos FIRMS con coincidencia de reportes de la UNGRD, reproyecci√≥n CRS. Construcci√≥n de zona de exclusi√≥n mediante buffer de 1 km alrededor de los puntos calientes, construendo una sola geometria con esto; se resta esta zona de exclusi√≥n al √°rea de interes, dando como resultado la regi√≥n donde se pueden ubicar los puntos no calientes aleatorios. Se generan 961 los puntos aleatorios dentro de la geometria previamente definida. Se construye el geodataframe de salida, evitando sobrescritura y exportaci√≥n final.

```python
import os
from src.random_absence_points import setup_logger, run_random_absence_sampling, format_report

ruta_aoi = r"D:\Maestria_Geomatica\Semestre_IV\Tesis_Python\Datos\SHP\AOI_25_09_19.shp"
ruta_firms = r"D:\Maestria_Geomatica\Semestre_IV\Puntos_unificados_25_11_05\Puntos_unificados_26_02_23\FIRMS_vs_UNGRD_coincidencias_26_02_23.shp"
```

```python
ruta_firms2 = r"D:\Maestria_Geomatica\Semestre_IV\Puntos_unificados_25_11_05\Puntos_unificados_26_02_23\FIRMS_vs_UNGRD_coincidencias_26_02_23_v4.shp"
carpeta_salida = r"D:\Maestria_Geomatica\Semestre_IV\Puntos_unificados_25_11_05\Puntos_unificados_26_02_23"
out_shp = os.path.join(carpeta_salida, "puntos_no_calientes_aleatorios_961_26_02_23.shp")

log_path = os.path.join(carpeta_salida, "log_puntos_no_calientes.txt")
logger = setup_logger(log_path)

summary = run_random_absence_sampling(
    aoi_path=ruta_aoi,
    points_path=ruta_firms2,
    out_path=out_shp,
    crs_target="EPSG:9377",
    buffer_m=1000,
    n_points=961,
    seed=42,
    max_attempts_factor=500,
    logger=logger
)

print(format_report(summary))
```
