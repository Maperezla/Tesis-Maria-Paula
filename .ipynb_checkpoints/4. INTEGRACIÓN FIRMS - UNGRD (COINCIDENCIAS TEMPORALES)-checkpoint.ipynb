## 4.2 Cruzar Bases de Datos de FIRMS y la UNGRD por Municipio y Fecha Buscando Coincidencia  ##

Integra dos fuentes de informaci√≥n sobre incendios: UNGRD (registros oficiales) en Excel, filtrados por departamentos de inter√©s y con fecha estructurada. FIRMS (detecciones satelitales) en shapefile, previamente asignadas a municipio y con fecha desglosada.

El objetivo es comparar FIRMS contra UNGRD y clasificar, para cada punto FIRMS, si existe: Coincidencia exacta (misma fecha) con un registro UNGRD del mismo municipio y departamento; Coincidencia en ventana temporal ¬±5 d√≠as (misma entidad territorial, fecha cercana) y Sin coincidencia (no existe registro UNGRD asociado en municipio/departamento o la fecha es inexistente).

El resultado es un shapefile enriquecido que conserva la geometr√≠a FIRMS y a√±ade campos de coincidencia.

### 4.2.1 Parte 1 ###

Normalizaci√≥n de texto, lectura y validaci√≥n del excel de la UNGRD; estandarizaci√≥n de municipio y departamento. Lectura y validaci√≥n del shapefile FIRMS, construcci√≥n de la fecha y estandarizaci√≥n de municipio y departamento. Preparaci√≥n de subconjutos para cruce. Elabora una uni√≥n entre UNGRD y FIRMS por municipio y departamento. Se calcula cu√°ntos d√≠as separan la detecci√≥n FIRMS de la fecha UNGRD y se crea la diferencia absoluta para escoger el match m√°s cercano. Se selecciona la coincidencia m√°s cercana, garantizando un unico match por puntos FIRMS, para un evento reportado por la UNGRD m√°s cercano emporalmente en su municipio y departamento. Se clasifican las coincidencias como exacta, centana de 5 d√≠as y sin coincidencia. Une la clasificaci√≥n al geodataframe FIRMS incorporandose las geometrias FIRMS, exportandose el shapefile de salida. Resumen final con coincidencias.
