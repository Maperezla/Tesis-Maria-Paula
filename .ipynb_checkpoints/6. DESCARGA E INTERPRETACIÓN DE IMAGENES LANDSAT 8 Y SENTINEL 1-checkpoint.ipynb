Se definen funciones auxiliares que permiten extraer el periodo de analisis a partir del nombre del archivo, se limpian valores NoData e infinitos y se calculan estadisticas descriptivas por bandas

````python
def extraer_info_nombre(nombre_archivo):
    """
    Extrae satélite, año y periodo desde nombres tipo:
    L8_2021_sep_oct.tif

    Retorna:
    - satelite: L8
    - anio: 2021
    - periodo: sep_oct
    """
    stem = Path(nombre_archivo).stem

    # Patrón esperado: L8_2021_sep_oct
    patron = r"^(?P<satelite>L8)_(?P<anio>\d{4})_(?P<periodo>.+)$"
    match = re.match(patron, stem)

    if match:
        satelite = match.group("satelite")
        anio = int(match.group("anio"))
        periodo = match.group("periodo")
    else:
        # Si el nombre no cumple el patrón, se conservan valores informativos
        satelite = None
        anio = None
        periodo = stem

    return satelite, anio, periodo


def limpiar_valores_banda(array_banda, nodata=None):
    """
    Convierte la banda a float64 y elimina valores no válidos:
    - NoData
    - NaN
    - infinitos positivos
    - infinitos negativos

    Retorna:
    - valores_validos: arreglo 1D con valores válidos
    - mascara_valida: matriz booleana True donde el píxel es válido
    """
    arr = array_banda.astype("float64")

    mascara_valida = np.isfinite(arr)

    if nodata is not None:
        mascara_valida &= arr != nodata

    valores_validos = arr[mascara_valida]

    return valores_validos, mascara_valida


def calcular_estadisticas_banda(array_banda, nodata=None):
    """
    Calcula estadísticas descriptivas para una banda.
    Ignora NoData, NaN e infinitos.
    """
    pixeles_totales = array_banda.size

    valores_validos, mascara_valida = limpiar_valores_banda(array_banda, nodata=nodata)

    pixeles_validos = valores_validos.size
    pixeles_nodata = pixeles_totales - pixeles_validos
    porcentaje_nodata = (pixeles_nodata / pixeles_totales) * 100 if pixeles_totales > 0 else np.nan

    if pixeles_validos == 0:
        return {
            "pixeles_totales": pixeles_totales,
            "pixeles_validos": pixeles_validos,
            "pixeles_nodata": pixeles_nodata,
            "porcentaje_nodata": porcentaje_nodata,
            "minimo": np.nan,
            "maximo": np.nan,
            "media": np.nan,
            "mediana": np.nan,
            "desviacion_estandar": np.nan,
            "P25": np.nan,
            "P75": np.nan,
            "IQR": np.nan,
            "coeficiente_variacion": np.nan,
        }

    minimo = np.min(valores_validos)
    maximo = np.max(valores_validos)
    media = np.mean(valores_validos)
    mediana = np.median(valores_validos)
    desviacion_estandar = np.std(valores_validos)
    p25 = np.percentile(valores_validos, 25)
    p75 = np.percentile(valores_validos, 75)
    iqr = p75 - p25

    # Coeficiente de variación: se calcula si la media no es cero
    if media != 0:
        coeficiente_variacion = desviacion_estandar / abs(media)
    else:
        coeficiente_variacion = np.nan

    return {
        "pixeles_totales": pixeles_totales,
        "pixeles_validos": pixeles_validos,
        "pixeles_nodata": pixeles_nodata,
        "porcentaje_nodata": porcentaje_nodata,
        "minimo": minimo,
        "maximo": maximo,
        "media": media,
        "mediana": mediana,
        "desviacion_estandar": desviacion_estandar,
        "P25": p25,
        "P75": p75,
        "IQR": iqr,
        "coeficiente_variacion": coeficiente_variacion,
    }


print("Funciones auxiliares definidas correctamente.")
````

## 6.1.2 Parte 2 ##

Estadisticas descriptivas de Landsat 8. Se calculan las estadísticas descriptivas donde se lee cada imagen una banda a la vez

````python
registros_estadisticos = []
registros_imagenes = []
errores = []

for archivo in archivos_tif:
    print(f"Procesando: {archivo.name}")

    try:
        satelite, anio, periodo = extraer_info_nombre(archivo.name)

        with rasterio.open(archivo) as src:
            numero_bandas = src.count
            ancho = src.width
            alto = src.height
            crs = str(src.crs)
            resolucion_x = src.res[0]
            resolucion_y = src.res[1]
            bounds = src.bounds
            nodata = src.nodata
            tipo_dato = src.dtypes[0] if len(src.dtypes) > 0 else None

            porcentajes_nodata_imagen = []
            pixeles_validos_imagen = []

            for banda_nombre in bandas_a_analizar:
                indice_banda = indice_por_banda[banda_nombre]

                # Validar que la imagen tenga esa banda
                if indice_banda > numero_bandas:
                    errores.append({
                        "archivo": archivo.name,
                        "error": f"La banda {banda_nombre} requiere índice {indice_banda}, pero la imagen solo tiene {numero_bandas} bandas."
                    })
                    continue

                # Leer solamente la banda requerida
                array_banda = src.read(indice_banda)

                estadisticas = calcular_estadisticas_banda(array_banda, nodata=nodata)

                porcentajes_nodata_imagen.append(estadisticas["porcentaje_nodata"])
                pixeles_validos_imagen.append(estadisticas["pixeles_validos"])

                registro = {
                    "archivo": archivo.name,
                    "satelite": satelite,
                    "anio": anio,
                    "periodo": periodo,
                    "banda": banda_nombre,
                    "indice_banda": indice_banda,
                    "ancho": ancho,
                    "alto": alto,
                    "CRS": crs,
                    "resolucion_x": resolucion_x,
                    "resolucion_y": resolucion_y,
                    "bounds": str(bounds),
                    "NoData": nodata,
                    "tipo_dato": tipo_dato,
                }

                registro.update(estadisticas)
                registros_estadisticos.append(registro)

            registros_imagenes.append({
                "archivo": archivo.name,
                "satelite": satelite,
                "anio": anio,
                "periodo": periodo,
                "numero_bandas": numero_bandas,
                "ancho": ancho,
                "alto": alto,
                "CRS": crs,
                "resolucion_x": resolucion_x,
                "resolucion_y": resolucion_y,
                "bounds": str(bounds),
                "NoData": nodata,
                "tipo_dato": tipo_dato,
                "porcentaje_nodata_promedio": np.nanmean(porcentajes_nodata_imagen) if len(porcentajes_nodata_imagen) > 0 else np.nan,
                "pixeles_validos_promedio": np.nanmean(pixeles_validos_imagen) if len(pixeles_validos_imagen) > 0 else np.nan,
                "bandas_analizadas": ", ".join(bandas_a_analizar),
            })

    except Exception as e:
        errores.append({
            "archivo": archivo.name,
            "error": str(e)
        })
        print(f"Error procesando {archivo.name}: {e}")

print("Proceso finalizado.")
print(f"Registros estadísticos generados: {len(registros_estadisticos)}")
print(f"Imágenes resumidas: {len(registros_imagenes)}")
print(f"Errores registrados: {len(errores)}")
````

Se contruyen los DataFrames con las bandas disponibles y analizadas 

````python
df_estadisticas = pd.DataFrame(registros_estadisticos)
df_resumen_imagenes = pd.DataFrame(registros_imagenes)
df_errores = pd.DataFrame(errores)

# Tabla de bandas disponibles y analizadas
df_bandas = pd.DataFrame([
    {
        "indice_banda": idx,
        "banda": nombre,
        "analizada": nombre in bandas_a_analizar
    }
    for idx, nombre in bandas_l8.items()
])

# Ordenar resultados para facilitar lectura
if not df_estadisticas.empty:
    df_estadisticas = df_estadisticas.sort_values(
        by=["anio", "periodo", "archivo", "indice_banda"],
        na_position="last"
    ).reset_index(drop=True)

if not df_resumen_imagenes.empty:
    df_resumen_imagenes = df_resumen_imagenes.sort_values(
        by=["anio", "periodo", "archivo"],
        na_position="last"
    ).reset_index(drop=True)

print("DataFrames construidos.")
print(f"df_estadisticas: {df_estadisticas.shape}")
print(f"df_resumen_imagenes: {df_resumen_imagenes.shape}")
print(f"df_bandas: {df_bandas.shape}")
print(f"df_errores: {df_errores.shape}")
````

### 6.2.1 Mosaico Sentinel 1 ###

Inicializa GEE; construye el área de interes como ee.Geometry desde shapefile y define parámetros globales (año, órbita, tamaño de lote); define funciones de preprocesamiento SAR con orbita ascendente. Construcción de la colección Sentinel 1, definiendose la selección de escenas Sentienl 1 que intersectan el área de interes, estan dentro del año, son IW GRD, cuentan con polarizaciones VV y VH y para la orbita previamente definida (ascendente). Se exporta CSV con los metadatos de las imagenes. Los páramentros de los lotes de descarga se encuentran en .json.

El sistema de logging crea o actualiza `logs/run.log`. Este archivo deja evidencia del inicio de la corrida, la configuración utilizada, el AOI cargado, las bandas finales y las tareas exportadas.

`window_days` controla ventanas temporales por días, no meses; `relative_orbit` permite restringir la geometría de adquisición; `eps` evita divisiones por cero; y `kernel_size` define la ventana espacial del filtro speckle.

````python
# =========================
# 9) Funciones auxiliares de depuración
# =========================
def show_bands(col, label):
    size = col.size().getInfo()
    print(f"{label} -> size: {size}")

    if size > 0:
        img = ee.Image(col.first())
        print(f"{label} -> bands: {img.bandNames().getInfo()}")
    else:
        print(f"{label} -> colección vacía")


def show_first_date(col, label):
    size = col.size().getInfo()

    if size > 0:
        img = ee.Image(col.first())
        date_str = ee.Date(img.get("system:time_start")).format("YYYY-MM-dd").getInfo()
        print(f"{label} -> primera fecha: {date_str}")
    else:
        print(f"{label} -> sin fechas (colección vacía)")


def show_sample_properties(col, label, props=None):
    size = col.size().getInfo()

    if size == 0:
        print(f"{label} -> colección vacía")
        return

    if props is None:
        props = [
            "fecha_id",
            "conteo_imagenes",
            "window_start",
            "window_end",
            "window_days",
            "aggregation",
        ]

    img = ee.Image(col.first())

    values = {}

    for p in props:
        try:
            values[p] = img.get(p).getInfo()
        except Exception:
            values[p] = None

    print(f"{label} -> propiedades muestra:")

    for k, v in values.items():
        print(f"   {k}: {v}")


def collection_to_feature_df(col: ee.ImageCollection, max_items: int = 500) -> pd.DataFrame:
    """
    Convierte una ee.ImageCollection de Sentinel-1 en una tabla de propiedades
    para diagnóstico de fechas, órbitas, slices y polarizaciones.
    """
    n = col.size().getInfo()
    n_use = min(n, max_items)

    if n_use == 0:
        return pd.DataFrame()

    img_list = col.toList(n_use)
    rows = []

    for i in range(n_use):
        img = ee.Image(img_list.get(i))

        props = img.toDictionary([
            "system:index",
            "system:time_start",
            "orbitProperties_pass",
            "relativeOrbitNumber_start",
            "sliceNumber",
            "platform_number",
            "instrumentMode",
            "productType",
            "transmitterReceiverPolarisation",
            "resolution_meters",
            "resolution",
            "orbitNumber_start",
        ]).getInfo()

        rows.append({
            "idx": i,
            "system_index": props.get("system:index"),
            "system_time_start": props.get("system:time_start"),
            "date": pd.to_datetime(props.get("system:time_start"), unit="ms", errors="coerce"),
            "orbit_pass": props.get("orbitProperties_pass"),
            "relative_orbit": props.get("relativeOrbitNumber_start"),
            "slice_number": props.get("sliceNumber"),
            "platform": props.get("platform_number"),
            "instrument_mode": props.get("instrumentMode"),
            "product_type": props.get("productType"),
            "polarizations": str(props.get("transmitterReceiverPolarisation")),
            "resolution_meters": props.get("resolution_meters"),
            "resolution": props.get("resolution"),
            "orbit_number_start": props.get("orbitNumber_start"),
        })

    df = pd.DataFrame(rows)

    if not df.empty:
        df = df.sort_values("date").reset_index(drop=True)

    return df


def inspect_mosaic_windows(
    col_raw: ee.ImageCollection,
    col_mosaic: ee.ImageCollection,
    max_mosaics: int = 20,
) -> pd.DataFrame:
    """
    Inspecciona las ventanas temporales usadas para construir los mosaicos.
    Esta función se mantiene porque ayuda a validar que los mosaicos se siguen
    generando con window_days.
    """
    n = min(col_mosaic.size().getInfo(), max_mosaics)

    if n == 0:
        return pd.DataFrame()

    mosaic_list = col_mosaic.toList(n)
    rows = []

    for i in range(n):
        mosaic = ee.Image(mosaic_list.get(i))

        window_start = mosaic.get("window_start").getInfo()
        window_end = mosaic.get("window_end").getInfo()
        fecha_id = mosaic.get("fecha_id").getInfo()
        conteo_imagenes = mosaic.get("conteo_imagenes").getInfo()

        subset = col_raw.filterDate(window_start, window_end)
        df_subset = collection_to_feature_df(subset, max_items=100)

        rel_orbits = (
            sorted(df_subset["relative_orbit"].dropna().unique().tolist())
            if not df_subset.empty and "relative_orbit" in df_subset.columns
            else []
        )

        slices = (
            sorted(df_subset["slice_number"].dropna().unique().tolist())
            if not df_subset.empty and "slice_number" in df_subset.columns
            else []
        )

        rows.append({
            "mosaic_idx": i,
            "fecha_id": fecha_id,
            "window_start": window_start,
            "window_end": window_end,
            "conteo_imagenes": conteo_imagenes,
            "n_relative_orbits": len(rel_orbits),
            "relative_orbits": rel_orbits,
            "n_slices": len(slices),
            "slices": slices,
        })

    return pd.DataFrame(rows)
````

Se verifica el efecto de cada etapa: colección base, mosaicos temporales, máscara de bordes, conversión a Gamma0, corrección topográfica, filtro speckle y cálculo de razón VV/VH.

La validación por etapas permite identificar en qué punto se pierden bandas, fechas, propiedades o píxeles válidos.

````python
# =========================
# 10) Construcción paso a paso para validación
# =========================
print("\n=========================")
print("VALIDACIÓN PASO A PASO")
print("=========================\n")


# -------------------------
# 10.1 Colección base
# -------------------------
if relative_orbit is None:
    col0 = build_collection_by_pass(
        aoi=aoi,
        year=year,
        orbit_pass=orbit,
        start_date=start_date,
        end_date=end_date,
    )
else:
    col0 = build_collection_by_pass_and_relative_orbit(
        aoi=aoi,
        year=year,
        orbit_pass=orbit,
        relative_orbit=int(relative_orbit),
        start_date=start_date,
        end_date=end_date,
    )

show_bands(col0, "0) build_collection")
show_first_date(col0, "0) build_collection")


# -------------------------
# 10.1.1 Diagnóstico de colección base
# -------------------------
df_col0 = collection_to_feature_df(col0, max_items=500)

print(df_col0.head(20))
print("Total imágenes en col0:", len(df_col0))

if not df_col0.empty:
    summary_pass = (
        df_col0.groupby("orbit_pass", dropna=False)
        .agg(
            n_scenes=("system_index", "count"),
            n_relative_orbits=("relative_orbit", "nunique"),
            n_slices=("slice_number", "nunique"),
            first_date=("date", "min"),
            last_date=("date", "max"),
        )
        .reset_index()
    )

    summary_rel_orbit = (
        df_col0.groupby("relative_orbit", dropna=False)
        .agg(
            n_scenes=("system_index", "count"),
            n_slices=("slice_number", "nunique"),
            first_date=("date", "min"),
            last_date=("date", "max"),
        )
        .reset_index()
        .sort_values("n_scenes", ascending=False)
    )

    summary_pass_rel = (
        df_col0.groupby(["orbit_pass", "relative_orbit"], dropna=False)
        .agg(
            n_scenes=("system_index", "count"),
            n_slices=("slice_number", "nunique"),
            first_date=("date", "min"),
            last_date=("date", "max"),
        )
        .reset_index()
        .sort_values(["orbit_pass", "n_scenes"], ascending=[True, False])
    )

    summary_slice = (
        df_col0.groupby("slice_number", dropna=False)
        .agg(
            n_scenes=("system_index", "count"),
            n_relative_orbits=("relative_orbit", "nunique"),
            first_date=("date", "min"),
            last_date=("date", "max"),
        )
        .reset_index()
        .sort_values("n_scenes", ascending=False)
    )

    print("\nResumen por pase")
    print(summary_pass)

    print("\nResumen por órbita relativa")
    print(summary_rel_orbit)

    print("\nResumen por pase + órbita relativa")
    print(summary_pass_rel)

    print("\nResumen por slice")
    print(summary_slice)
else:
    summary_pass = pd.DataFrame()
    summary_rel_orbit = pd.DataFrame()
    summary_pass_rel = pd.DataFrame()
    summary_slice = pd.DataFrame()

    print("La colección base está vacía. No se generan resúmenes.")
````

````python
# -------------------------
# 10.2 Mosaicos temporales
# -------------------------
col1 = build_temporal_mosaics(
    col=col0,
    aoi=aoi,
    start_date=start_date,
    end_date=end_date,
    window_days=window_days,
    reducer=mosaic_reducer,
)

show_bands(col1, "1) build_temporal_mosaics")
show_first_date(col1, "1) build_temporal_mosaics")
show_sample_properties(col1, "1) build_temporal_mosaics")

df_windows = inspect_mosaic_windows(col0, col1, max_mosaics=20)
print(df_windows)


# -------------------------
# 10.3 Máscara de borde
# -------------------------
col2 = col1.map(border_noise_mask)

show_bands(col2, "2) + border_noise_mask")
show_first_date(col2, "2) + border_noise_mask")
show_sample_properties(col2, "2) + border_noise_mask")


# -------------------------
# 10.4 Gamma0
# -------------------------
col3 = col2.map(gamma0_db)

show_bands(col3, "3) + gamma0_db")
show_first_date(col3, "3) + gamma0_db")
show_sample_properties(col3, "3) + gamma0_db")


# -------------------------
# 10.5 Terrain flattening
# -------------------------
dem = ee.Image(dem_id)

col4 = col3.map(lambda img: terrain_flattening(img, dem=dem))

show_bands(col4, "4) + terrain_flattening")
show_first_date(col4, "4) + terrain_flattening")
show_sample_properties(col4, "4) + terrain_flattening")


# -------------------------
# 10.6 Speckle
# -------------------------
col5 = col4.map(lambda img: refined_lee_spatial(img, kernel_size=kernel_size))

show_bands(col5, "5) + refined_lee_spatial")
show_first_date(col5, "5) + refined_lee_spatial")
show_sample_properties(col5, "5) + refined_lee_spatial")


# -------------------------
# 10.7 Ratio VV/VH
# -------------------------
col6 = col5.map(lambda img: add_ratio_db(img, eps=eps))

show_bands(col6, "6) + add_ratio_db")
show_first_date(col6, "6) + add_ratio_db")
show_sample_properties(col6, "6) + add_ratio_db")


# -------------------------
# 10.8 Flujo base SIN pre/post
# -------------------------
col7 = col6.sort("system:time_start")

show_bands(col7, "7) flujo base SIN pre/post")
show_first_date(col7, "7) flujo base SIN pre/post")
show_sample_properties(
    col7,
    "7) flujo base SIN pre/post",
    props=[
        "fecha_id",
        "conteo_imagenes",
        "window_start",
        "window_end",
        "window_days",
        "aggregation",
    ],
)

first_img_col7 = ee.Image(col7.first())

print("Bandas col7:", first_img_col7.bandNames().getInfo())

for p in [
    "fecha_id",
    "conteo_imagenes",
    "window_start",
    "window_end",
    "window_days",
    "aggregation",
]:
    try:
        print(p, ":", first_img_col7.get(p).getInfo())
    except Exception:
        print(p, ": <no disponible>")
````

Se revisan propiedades temporales de la primera imagen final, como `fecha_id`, `window_start`, `window_end`, `window_days` y `aggregation`. Estas propiedades son fundamentales para documentar cómo se construyó cada mosaico temporal.

El flujo base exporta únicamente `VV`, `VH`, `angle` y `VVVH_ratio`. Además, valida que no estén presentes bandas de cambio pre/post: `VV_Difference`, `VH_Difference` y `VHVV_Difference`.

La función `export_batch_to_drive()` exporta subconjuntos de imágenes de una `ee.ImageCollection` a Google Drive. El uso de `batch_start` y `batch_size` permite controlar cuántas tareas se lanzan simultáneamente.

````python
# =========================
# 11) Pipeline completo
# =========================
print("\n=========================")
print("PIPELINE COMPLETO")
print("=========================\n")

# Uso explícito del flujo base.
col_final = build_pipeline_base(aoi, cfg)

n_final = col_final.size().getInfo()
print("N imágenes finales:", n_final)

if n_final == 0:
    raise ValueError(
        "La colección final quedó vacía. "
        "Revisa AOI, fechas, órbita, órbita relativa o window_days."
    )

first_img = ee.Image(col_final.first())
bands_final = first_img.bandNames().getInfo()

print("Bandas finales:", bands_final)
print(
    "Primera fecha final:",
    ee.Date(first_img.get("system:time_start")).format("YYYY-MM-dd").getInfo(),
)

logger.info(f"N imágenes finales: {n_final}")
logger.info(f"Bandas finales: {bands_final}")


# =========================
# 12) Verificación adicional de propiedades
# =========================
print("\n=========================")
print("PROPIEDADES DE LA PRIMERA IMAGEN FINAL")
print("=========================\n")

props_to_check = [
    "fecha_id",
    "conteo_imagenes",
    "window_start",
    "window_end",
    "window_days",
    "aggregation",
]

for p in props_to_check:
    try:
        print(f"{p}: {first_img.get(p).getInfo()}")
    except Exception:
        print(f"{p}: <no disponible>")


# =========================
# 13) Lista de bandas a exportar
# =========================
EXPORT_BANDS = [
    "VV",
    "VH",
    "angle",
    "VVVH_ratio",
]

print("\nEXPORT_BANDS:", EXPORT_BANDS)

missing_bands = [b for b in EXPORT_BANDS if b not in bands_final]

if missing_bands:
    raise ValueError(f"Faltan bandas para exportar: {missing_bands}")

forbidden_bands = [
    "VV_Difference",
    "VH_Difference",
    "VHVV_Difference",
]

present_forbidden_bands = [b for b in forbidden_bands if b in bands_final]

if present_forbidden_bands:
    raise ValueError(
        "Aún aparecen bandas pre/post en la colección final: "
        f"{present_forbidden_bands}"
    )

print("Validación OK: no hay bandas pre/post en la colección final.")


# =========================
# 14) Función de exportación por lotes
# =========================
def export_batch_to_drive(
    col,
    aoi,
    year,
    folder,
    scale,
    max_pixels,
    batch_start=0,
    batch_size=10,
    export_bands=None,
):
    """
    Exporta por lotes imágenes de una ee.ImageCollection a Google Drive.

    Parámetros
    ----------
    col : ee.ImageCollection
        Colección final de mosaicos Sentinel-1.
    aoi : ee.Geometry
        Área de interés para clip y región de exportación.
    year : int
        Año de análisis.
    folder : str
        Carpeta en Google Drive.
    scale : int | float
        Resolución espacial de exportación.
    max_pixels : int | float
        Máximo de píxeles permitido por GEE.
    batch_start : int
        Índice inicial del lote.
    batch_size : int
        Número de imágenes a exportar.
    export_bands : list[str] | None
        Bandas a exportar.

    Retorna
    -------
    list
        Lista de tareas de exportación creadas.
    """
    if export_bands is None:
        export_bands = EXPORT_BANDS

    n = col.size().getInfo()
    print(f"[EXPORT] year={year} -> n_images={n}")

    if n == 0:
        print("[EXPORT] Colección vacía. No se crean tareas.")
        return []

    batch_end = min(batch_start + batch_size, n)

    if batch_start >= n:
        raise ValueError(f"batch_start={batch_start} >= n={n}")

    img_list = col.toList(n)
    tasks = []

    for i in range(batch_start, batch_end):
        img = ee.Image(img_list.get(i))

        try:
            fecha_id = img.get("fecha_id").getInfo()
        except Exception:
            fecha_id = f"{year}_{i}"

        desc = f"S1_{fecha_id}_{i}"

        out = img.select(export_bands).toFloat().clip(aoi)

        task = ee.batch.Export.image.toDrive(
            image=out,
            description=desc,
            folder=folder,
            fileNamePrefix=desc,
            region=aoi,
            scale=scale,
            maxPixels=max_pixels,
            fileFormat="GeoTIFF",
        )

        task.start()
        tasks.append(task)

        print(f"[EXPORT] Lanzada tarea {i}: {desc}")

    print(f"[EXPORT] Tareas lanzadas: {len(tasks)} ({batch_start}..{batch_end - 1})")
    print(f"[EXPORT] Siguiente batch_start sugerido: {batch_end}")

    return tasks


# =========================
# 15) Vista rápida de fechas disponibles
# =========================
print("\n=========================")
print("FECHAS DISPONIBLES")
print("=========================\n")

try:
    fechas = col_final.aggregate_array("fecha_id").getInfo()

    print("Primeras 20 fechas:")
    for f in fechas[:20]:
        print(" -", f)

except Exception as e:
    print("No fue posible extraer fechas:", e)


# =========================
# 16) Exportación
# =========================
DRIVE_FOLDER = "Imagenes_GEE"
EXPORT_SCALE = 30
EXPORT_MAX_PIXELS = 1e13
BATCH_START = 0
BATCH_SIZE = 5

print("\n=========================")
print("EXPORTACIÓN")
print("=========================\n")

print("Carpeta Drive:", DRIVE_FOLDER)
print("Scale:", EXPORT_SCALE)
print("Max pixels:", EXPORT_MAX_PIXELS)
print("Batch start:", BATCH_START)
print("Batch size:", BATCH_SIZE)

tasks = export_batch_to_drive(
    col=col_final,
    aoi=aoi,
    year=year,
    folder=DRIVE_FOLDER,
    scale=EXPORT_SCALE,
    max_pixels=EXPORT_MAX_PIXELS,
    batch_start=BATCH_START,
    batch_size=BATCH_SIZE,
    export_bands=EXPORT_BANDS,
)

print(f"\nSe lanzaron {len(tasks)} tareas.")

logger.info(f"Tareas exportadas: {len(tasks)}")
logger.info("=== END RUN ===")


# =========================
# 17) Validaciones finales
# =========================
print("\n=========================")
print("VALIDACIONES FINALES")
print("=========================\n")

print("Bandas finales:", bands_final)

for p in [
    "fecha_id",
    "window_start",
    "window_end",
    "window_days",
    "aggregation",
]:
    try:
        print(p, ":", first_img.get(p).getInfo())
    except Exception:
        print(p, ": <no disponible>")


# Comparación de proyecciones y geometrías
img_mosaic = ee.Image(col1.first())
img_final = ee.Image(col_final.first())

print("MOSAIC projection:", img_mosaic.select("VV").projection().getInfo())
print("FINAL projection:", img_final.select("VV").projection().getInfo())

print("MOSAIC geometry:", img_mosaic.geometry().bounds().getInfo())
print("FINAL geometry:", img_final.geometry().bounds().getInfo())


def valid_pixel_count(img, band, region, scale=30):
    """
    Cuenta píxeles válidos de una banda dentro del AOI.
    Útil para identificar en qué etapa se pierden píxeles por máscara.
    """
    stats = img.select(band).mask().reduceRegion(
        reducer=ee.Reducer.sum(),
        geometry=region,
        scale=scale,
        maxPixels=1e13,
    ).getInfo()

    return stats


imgs = {
    "mosaic": ee.Image(col1.first()),
    "border": ee.Image(col2.first()),
    "gamma0": ee.Image(col3.first()),
    "terrain": ee.Image(col4.first()),
    "lee": ee.Image(col5.first()),
    "ratio": ee.Image(col6.first()),
    "final": ee.Image(col7.first()),
}

for name, img in imgs.items():
    try:
        print(f"\n{name}")
        print("VV:", valid_pixel_count(img, "VV", aoi))
    except Exception as e:
        print(f"{name} -> error: {e}")


# =========================
# 18) Exportar CSV diagnósticos
# =========================
print("\n=========================")
print("EXPORTACIÓN DE CSV DIAGNÓSTICOS")
print("=========================\n")

df_col0 = collection_to_feature_df(col0, max_items=500)

print(df_col0.head(20))
print("\nTotal imágenes:", len(df_col0))

if not df_col0.empty:
    chronological_table = df_col0[
        [
            "date",
            "orbit_pass",
            "relative_orbit",
            "slice_number",
            "platform",
            "system_index",
            "polarizations",
        ]
    ].sort_values("date").reset_index(drop=True)

    print(chronological_table.head(20))
else:
    chronological_table = pd.DataFrame()
    print("No se puede generar tabla cronológica porque df_col0 está vacío.")

out_dir = r"D:\Maestria_Geomatica\Semestre_IV\Tesis_Python\Datos\XLS\outputs_diagnostico"
os.makedirs(out_dir, exist_ok=True)

df_col0.to_csv(os.path.join(out_dir, "coleccion_detallada_base202201.csv"), index=False)
summary_pass.to_csv(os.path.join(out_dir, "resumen_por_pase_base202201.csv"), index=False)
summary_rel_orbit.to_csv(os.path.join(out_dir, "resumen_por_orbita_relativa_base202201.csv"), index=False)
summary_pass_rel.to_csv(os.path.join(out_dir, "resumen_por_pase_orbita_base202201.csv"), index=False)
summary_slice.to_csv(os.path.join(out_dir, "resumen_por_slice_base202201.csv"), index=False)
chronological_table.to_csv(os.path.join(out_dir, "tabla_cronologica_base202201.csv"), index=False)
df_windows.to_csv(os.path.join(out_dir, "resumen_ventanas_mosaico_base202201.csv"), index=False)

print("CSV exportados en:", out_dir)
````

`extraer_info_nombre()` intenta obtener satélite, año y periodo a partir del nombre del archivo. Actualmente el patrón documentado está orientado a nombres tipo `L8_2021_sep_oct.tif`, por lo que puede no extraer correctamente nombres Sentinel-1.

`limpiar_valores_banda()` elimina valores NoData, NaN e infinitos.

`calcular_estadisticas_banda()` calcula estadísticas descriptivas por banda: mínimo, máximo, media, mediana, desviación estándar, percentiles 25 y 75, IQR y coeficiente de variación.


````python
def extraer_info_nombre(nombre_archivo):
    """
    Extrae satélite, año y periodo desde nombres tipo:
    L8_2021_sep_oct.tif

    Retorna:
    - satelite: L8
    - anio: 2021
    - periodo: sep_oct
    """
    stem = Path(nombre_archivo).stem

    # Patrón esperado: L8_2021_sep_oct
    patron = r"^(?P<satelite>L8)_(?P<anio>\d{4})_(?P<periodo>.+)$"
    match = re.match(patron, stem)

    if match:
        satelite = match.group("satelite")
        anio = int(match.group("anio"))
        periodo = match.group("periodo")
    else:
        # Si el nombre no cumple el patrón, se conservan valores informativos
        satelite = None
        anio = None
        periodo = stem

    return satelite, anio, periodo


def limpiar_valores_banda(array_banda, nodata=None):
    """
    Convierte la banda a float64 y elimina valores no válidos:
    - NoData
    - NaN
    - infinitos positivos
    - infinitos negativos

    Retorna:
    - valores_validos: arreglo 1D con valores válidos
    - mascara_valida: matriz booleana True donde el píxel es válido
    """
    arr = array_banda.astype("float64")

    mascara_valida = np.isfinite(arr)

    if nodata is not None:
        mascara_valida &= arr != nodata

    valores_validos = arr[mascara_valida]

    return valores_validos, mascara_valida


def calcular_estadisticas_banda(array_banda, nodata=None):
    """
    Calcula estadísticas descriptivas para una banda.
    Ignora NoData, NaN e infinitos.
    """
    pixeles_totales = array_banda.size

    valores_validos, mascara_valida = limpiar_valores_banda(array_banda, nodata=nodata)

    pixeles_validos = valores_validos.size
    pixeles_nodata = pixeles_totales - pixeles_validos
    porcentaje_nodata = (pixeles_nodata / pixeles_totales) * 100 if pixeles_totales > 0 else np.nan

    if pixeles_validos == 0:
        return {
            "pixeles_totales": pixeles_totales,
            "pixeles_validos": pixeles_validos,
            "pixeles_nodata": pixeles_nodata,
            "porcentaje_nodata": porcentaje_nodata,
            "minimo": np.nan,
            "maximo": np.nan,
            "media": np.nan,
            "mediana": np.nan,
            "desviacion_estandar": np.nan,
            "P25": np.nan,
            "P75": np.nan,
            "IQR": np.nan,
            "coeficiente_variacion": np.nan,
        }

    minimo = np.min(valores_validos)
    maximo = np.max(valores_validos)
    media = np.mean(valores_validos)
    mediana = np.median(valores_validos)
    desviacion_estandar = np.std(valores_validos)
    p25 = np.percentile(valores_validos, 25)
    p75 = np.percentile(valores_validos, 75)
    iqr = p75 - p25

    # Coeficiente de variación: se calcula si la media no es cero
    if media != 0:
        coeficiente_variacion = desviacion_estandar / abs(media)
    else:
        coeficiente_variacion = np.nan

    return {
        "pixeles_totales": pixeles_totales,
        "pixeles_validos": pixeles_validos,
        "pixeles_nodata": pixeles_nodata,
        "porcentaje_nodata": porcentaje_nodata,
        "minimo": minimo,
        "maximo": maximo,
        "media": media,
        "mediana": mediana,
        "desviacion_estandar": desviacion_estandar,
        "P25": p25,
        "P75": p75,
        "IQR": iqr,
        "coeficiente_variacion": coeficiente_variacion,
    }


print("Funciones auxiliares definidas correctamente.")
````

````python
registros_estadisticos = []
registros_imagenes = []
errores = []

for archivo in archivos_tif:
    print(f"Procesando: {archivo.name}")

    try:
        satelite, anio, periodo = extraer_info_nombre(archivo.name)

        with rasterio.open(archivo) as src:
            numero_bandas = src.count
            ancho = src.width
            alto = src.height
            crs = str(src.crs)
            resolucion_x = src.res[0]
            resolucion_y = src.res[1]
            bounds = src.bounds
            nodata = src.nodata
            tipo_dato = src.dtypes[0] if len(src.dtypes) > 0 else None

            porcentajes_nodata_imagen = []
            pixeles_validos_imagen = []

            for banda_nombre in bandas_a_analizar:
                indice_banda = indice_por_banda[banda_nombre]

                # Validar que la imagen tenga esa banda
                if indice_banda > numero_bandas:
                    errores.append({
                        "archivo": archivo.name,
                        "error": f"La banda {banda_nombre} requiere índice {indice_banda}, pero la imagen solo tiene {numero_bandas} bandas."
                    })
                    continue

                # Leer solamente la banda requerida
                array_banda = src.read(indice_banda)

                estadisticas = calcular_estadisticas_banda(array_banda, nodata=nodata)

                porcentajes_nodata_imagen.append(estadisticas["porcentaje_nodata"])
                pixeles_validos_imagen.append(estadisticas["pixeles_validos"])

                registro = {
                    "archivo": archivo.name,
                    "satelite": satelite,
                    "anio": anio,
                    "periodo": periodo,
                    "banda": banda_nombre,
                    "indice_banda": indice_banda,
                    "ancho": ancho,
                    "alto": alto,
                    "CRS": crs,
                    "resolucion_x": resolucion_x,
                    "resolucion_y": resolucion_y,
                    "bounds": str(bounds),
                    "NoData": nodata,
                    "tipo_dato": tipo_dato,
                }

                registro.update(estadisticas)
                registros_estadisticos.append(registro)

            registros_imagenes.append({
                "archivo": archivo.name,
                "satelite": satelite,
                "anio": anio,
                "periodo": periodo,
                "numero_bandas": numero_bandas,
                "ancho": ancho,
                "alto": alto,
                "CRS": crs,
                "resolucion_x": resolucion_x,
                "resolucion_y": resolucion_y,
                "bounds": str(bounds),
                "NoData": nodata,
                "tipo_dato": tipo_dato,
                "porcentaje_nodata_promedio": np.nanmean(porcentajes_nodata_imagen) if len(porcentajes_nodata_imagen) > 0 else np.nan,
                "pixeles_validos_promedio": np.nanmean(pixeles_validos_imagen) if len(pixeles_validos_imagen) > 0 else np.nan,
                "bandas_analizadas": ", ".join(bandas_a_analizar),
            })

    except Exception as e:
        errores.append({
            "archivo": archivo.name,
            "error": str(e)
        })
        print(f"Error procesando {archivo.name}: {e}")

print("Proceso finalizado.")
print(f"Registros estadísticos generados: {len(registros_estadisticos)}")
print(f"Imágenes resumidas: {len(registros_imagenes)}")
print(f"Errores registrados: {len(errores)}")
````

````python
df_estadisticas = pd.DataFrame(registros_estadisticos)
df_resumen_imagenes = pd.DataFrame(registros_imagenes)
df_errores = pd.DataFrame(errores)

# Tabla de bandas disponibles y analizadas
df_bandas = pd.DataFrame([
    {
        "indice_banda": idx,
        "banda": nombre,
        "analizada": nombre in bandas_a_analizar
    }
    for idx, nombre in bandas_s1.items()
])

# Ordenar resultados para facilitar lectura
if not df_estadisticas.empty:
    df_estadisticas = df_estadisticas.sort_values(
        by=["anio", "periodo", "archivo", "indice_banda"],
        na_position="last"
    ).reset_index(drop=True)

if not df_resumen_imagenes.empty:
    df_resumen_imagenes = df_resumen_imagenes.sort_values(
        by=["anio", "periodo", "archivo"],
        na_position="last"
    ).reset_index(drop=True)

print("DataFrames construidos.")
print(f"df_estadisticas: {df_estadisticas.shape}")
print(f"df_resumen_imagenes: {df_resumen_imagenes.shape}")
print(f"df_bandas: {df_bandas.shape}")
print(f"df_errores: {df_errores.shape}")
````

### 6.2.3 Indices Sentinel 1 ###

Se generaron productos Sentinel-1 con índices de cambio temporal pre/post. Este procedimiento parte de GeoTIFF Sentinel-1 ya existentes en una carpeta local y genera nuevos archivos con bandas derivadas. Se producen imágenes Sentinel-1 que integren información original y métricas de cambio temporal.

El flujo está orientado a:

- Localizar el directorio raíz del proyecto.
- Importar funciones propias desde la carpeta `src/`.
- Definir carpetas locales de entrada y salida.
- Ejecutar la función `process_all_s1_prepost()`.
- Registrar el proceso en un archivo `.log`.
- Visualizar una tabla de trazabilidad del procesamiento.
- Verificar la estructura, metadatos y estadísticas básicas de un GeoTIFF generado

La función process_all_s1_prepost() procesa todos los archivos Sentinel-1 ubicados en `input_dir` y genera productos derivados en `output_dir`.


````python
from pathlib import Path
import logging
import sys

PROJECT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()

if str(PROJECT) not in sys.path:
    sys.path.insert(0, str(PROJECT))

from src.s1_prepost_indices import process_all_s1_prepost

input_dir = Path(r"D:\Maestria_Geomatica\Semestre_IV\Tesis_Python\Datos\TIFF\S1_V2")
output_dir = Path(r"D:\Maestria_Geomatica\Semestre_IV\Tesis_Python\Datos\TIFF\S1_V2_PREPOST")

output_dir.mkdir(parents=True, exist_ok=True)

log_path = output_dir / "s1_prepost_processing.log"

logging.basicConfig(
    filename=str(log_path),
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
    force=True
)

logger = logging.getLogger("S1_PREPOST")

trace = process_all_s1_prepost(
    input_dir=input_dir,
    output_dir=output_dir,
    eps=1e-10,
    nodata=-9999.0,
    logger=logger
)

display(trace)
display(trace["status"].value_counts(dropna=False))
````

Verificación del GeoTIFF resultante y verificación de metadatos yestadisticas básicas por banda

````python
import rasterio
from pathlib import Path
import numpy as np

output_dir = Path(r"D:\Maestria_Geomatica\Semestre_IV\Tesis_Python\Datos\TIFF\S1_V2_PREPOST")

tif = sorted(output_dir.glob("*_prepost_indices.tif"))[0]

with rasterio.open(tif) as src:
    print("Archivo:", tif.name)
    print("CRS:", src.crs)
    print("Width, height:", src.width, src.height)
    print("Bandas:", src.count)
    print("NoData:", src.nodata)
    print("Descriptions:", src.descriptions)

    for i in range(1, src.count + 1):
        arr = src.read(i).astype("float32")
        nodata = src.nodata
        arr = arr[arr != nodata]
        arr = arr[np.isfinite(arr)]

        print(
            i,
            src.descriptions[i - 1],
            "min:", arr.min() if arr.size else None,
            "max:", arr.max() if arr.size else None,
            "mean:", arr.mean() if arr.size else None,
        )
````
