## 1.1 Reproyectar Puntos Calientes de la Plataforma FIRMS para Colombia ##

Automatiza la reproyecci√≥n de m√∫ltiples shapefiles a un sistema de coordenadas est√°ndar: WGS84 (EPSG:4326), recorriendo recursivamente una carpeta principal y sus subcarpetas. El proceso evita sobrescribir archivos existentes creando una copia por cada shapefile. Produce un reporte final con el total procesado

### 1.1.1 Parte 1 ### 

Carga librer√≠as para sistema de archivos, SIG vectorial, logging y barra de progreso. Se define d√≥nde estan los shapefile a procesar. Se define d√≥nde estan los shapefile a procesar. Se crea un archivo log_reproyeccion.txt para registrar eventos. Se busca identificar si algun shapefile fall√≥ y por qu√©. Se inicializan metricas de ejecuci√≥n y se define el CRS de salida. Se recortan todos los archivos de la carpeta principal, excluyendose los que ya estan reproyectados y se guarda una copia de los mismos. Se reproyecta el AOI y se genera un reporte final 

```python
import os
from src.shp_reprojector import (
    setup_logger, batch_reproject_shps, reproject_one_shp, format_report, ReprojectSummary
)

shapefile_aoi = r"D:\Maestria_Geomatica\Semestre_IV\Tesis_Python\Datos\SHP\AOI_25_09_19.shp"
carpeta_principal = r"D:\Maestria_Geomatica\Semestre_IV\Puntos_calientes_Colombia_FIRMS_25_09_19"
target_crs = "EPSG:4326"

log_path = os.path.join(carpeta_principal, "log_reproyeccion_26_02_20.txt")
logger = setup_logger(log_path)

summary_main = batch_reproject_shps(
    root_folder=carpeta_principal,
    target_crs=target_crs,
    suffix="_wgs84",
    exclude_if_contains="_wgs84",
    version=True,
    logger=logger
)

# AOI aparte (tambi√©n versionado)
summary_aoi = ReprojectSummary(total=1)
try:
    out_aoi = reproject_one_shp(shapefile_aoi, target_crs, suffix="_wgs84", version=True, logger=logger)
    summary_aoi.ok = 1
    print(f"‚úÖ AOI reproyectado: {out_aoi}")
except Exception as e:
    summary_aoi.err = 1
    logger.error(f"No se pudo reproyectar el AOI - Error: {e}")
    print("‚ùå Error en AOI. Ver log.")
```

```python
# Reporte final consolidado
summary_all = ReprojectSummary(
    total=summary_main.total + summary_aoi.total,
    ok=summary_main.ok + summary_aoi.ok,
    err=summary_main.err + summary_aoi.err
)

reporte = format_report(summary_all)
print(reporte)
logger.info(reporte)
```
